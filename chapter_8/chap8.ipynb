```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi

def sample_and_plot(dimensions, num_samples=10000):
    # Sample from the standard normal distribution
    samples = np.random.normal(0, 1, (num_samples, dimensions))
    
    # Calculate the distances from the center
    distances = np.sqrt(np.sum(samples**2, axis=1))
    
    # Plot the histogram of distances
    plt.hist(distances, bins=30, density=True, alpha=0.6, label=f'{dimensions} dimensions')
    
    # Plot the theoretical chi distribution
    df = dimensions  # Degrees of freedom for the chi distribution
    x = np.linspace(chi.ppf(0.01, df), chi.ppf(0.99, df), 100)
    plt.plot(x, chi.pdf(x, df), 'r-', lw=2, label=f'Chi PDF {dimensions}D')
    
    plt.title(f'Distribution of Distances in {dimensions}-Dimensional Space')
    plt.xlabel('Distance from the Origin')
    plt.ylabel('Probability Density')
    plt.legend()

plt.figure(figsize=(14, 5))

# Plot for 25, 100, and 500 dimensions
plt.subplot(1, 3, 1)
sample_and_plot(25)

plt.subplot(1, 3, 2)
sample_and_plot(100)

plt.subplot(1, 3, 3)
sample_and_plot(500)

plt.tight_layout()
plt.show()
```
